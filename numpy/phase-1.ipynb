### List v/s Array

```python
py_List = [1,2,3]
result = [x*2 for x in py_List]
print("Python List: ", result)
```

```python
py_array = np.array([1,2,3,4])
print("Python List: ", py_array * 2) 
```

```python
import time
start = time.time()
py_list = [i*2 for i in range(10000)]
print("List ops time: ", time.time() - start)
```

```python
start = time.time()
np_array= np.arange(10000) * 2
print("Array ops time: ", time.time() - start)
```
