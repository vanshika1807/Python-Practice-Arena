## Series
#### It is like a coloumn / row in a table 

```python
import pandas as pd
marks = pd.Series([23,43,67,93,45,56])
print(marks)
```

```python
print(marks[2:6])
```

```python
marks = pd.Series([23,43,67,93,45,56], index = ['a','b','c','d','e','f'])
print(marks)
```

## Data Frame

#### A dataframe is like a table --> have both row & coloumns

```python
student = {
    'Marks' : [32,54,65,22],
    'Sub': ['A', 'B', 'C', 'D']
}

df = pd.DataFrame(student, index = [1,2,3,4])
print(df)
```

```python
print(df.iloc[[1,3]])
```

```python
print(df.loc[[1,3]])
```
